```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Создаем экземпляр драйвера (например, для Chrome)
driver = webdriver.Chrome()
```

```python
# Открываем страницу
driver.get("https://musescore.com/user/8927976/scores/1993341")
```

```python
pages = driver.find_elements(By.CSS_SELECTOR, ".A8huy")
for i in pages:
  ActionChains(driver).scroll_to_element(i).perform()
  time.sleep(1)
  svg_element = driver.find_element(By.CSS_SELECTOR, ".MHaWn")
  print(svg_element.get_attribute("src"))
```

```python
driver.quit()
```
